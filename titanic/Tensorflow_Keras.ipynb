```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# Load the data
df = pd.read_csv("train.csv")
```

```python
# Preprocess the data
df = df.drop(["Name", "Ticket", "Cabin"], axis=1)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df = pd.get_dummies(df, columns=["Pclass","Embarked","Sex"], drop_first=True)
df
```

```python
# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop("Survived", axis=1))
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(scaled_data, df["Survived"], test_size=0.2, shuffle=True, random_state=10)
```

```python
# Build the model
model = Sequential()
#model.add(Dense(128, activation="relu", input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation="relu", input_shape=(X_train.shape[1],)))
model.add(Dense(16, activation="relu"))
#model.add(Dense(32, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_test, y_test))
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
```

```python
accuracy = score[1]
print("Test accuracy:", accuracy)
```
